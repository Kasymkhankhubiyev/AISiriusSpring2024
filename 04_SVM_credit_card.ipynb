```python
import os
import sys

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# models
import statsmodels.api as sm
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier

# metrics
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import classification_report


# visualization
import seaborn as sns
import matplotlib.pyplot as plt

# paralleling
from joblib import parallel_backend 


RANDOM_SEED = 20240418
```

```python
PATH_TO_FILE = './SVM_creditcard.csv'
```

```python
df = pd.read_csv(PATH_TO_FILE).rename(columns={'Class':'Fraud'})
```

```python
FEATURES = [col for col in df.columns if col not in ['Time', 'Fraud']]
len(FEATURES)
```

```python
ncols = 4
nrows = int(np.ceil(len(FEATURES)/ncols))

fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(3*ncols, 3*nrows))

for idx, feature in enumerate(FEATURES):
    axs[idx//ncols][idx%ncols].set_title(feature)
    sns.histplot(df[feature], bins=20, kde=True, ax=axs[idx//ncols][idx%ncols])
plt.tight_layout()
```

```python
sns.heatmap(df[FEATURES].corr())
plt.show()
```

---

```python
X_train, X_test, Y_train, Y_test = train_test_split(df.drop(['Time','Fraud'], axis=1), df['Fraud'], test_size=0.2)
```

```python
bcl =  BaggingClassifier(
        SVC(
        kernel='linear',
        random_state=RANDOM_SEED
    ),
    random_state=RANDOM_SEED,
    max_samples=0.2,
    n_jobs=4
)
```

```python
bcl.fit(X_train, Y_train)
```

```python
bcl.predict(X_train)
```

```python
prediction = bcl.predict(X_train)
cm = confusion_matrix(Y_train, prediction)
displ = ConfusionMatrixDisplay(cm)
displ.plot()
plt.show()
print(classification_report(Y_train, prediction))
```

```python
prediction = bcl.predict(X_test)
cm = confusion_matrix(Y_test, prediction)
displ = ConfusionMatrixDisplay(cm)
displ.plot()
plt.show()
print(classification_report(Y_test, prediction))
```

---
## Run logistic regression

```python
reg = LogisticRegression(max_iter=10**4)

reg.fit(X_train, Y_train)
```

```python
prediction = reg.predict(X_train)
cm = confusion_matrix(Y_train, prediction)
displ = ConfusionMatrixDisplay(cm)
displ.plot()
plt.show()
print(classification_report(Y_train, prediction))
```

```python
prediction = reg.predict(X_test)
cm = confusion_matrix(Y_test, prediction)
displ = ConfusionMatrixDisplay(cm)
displ.plot()
plt.show()
print(classification_report(Y_test, prediction))
```

---
### Decision Tree

```python
dt = DecisionTreeClassifier(random_state=RANDOM_SEED,
                            criterion='gini',
                            max_depth=15,
                            min_samples_leaf=5)
```

```python
dt.fit(X_train, Y_train)
```

```python
prediction = dt.predict(X_train)
cm = confusion_matrix(Y_train, prediction)
displ = ConfusionMatrixDisplay(cm)
displ.plot()
plt.show()
print(classification_report(Y_train, prediction))
```

```python
prediction = dt.predict(X_test)
cm = confusion_matrix(Y_test, prediction)
displ = ConfusionMatrixDisplay(cm)
displ.plot()
plt.show()
print(classification_report(Y_test, prediction))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(random_state=RANDOM_SEED,
                            max_depth=15,
                            min_samples_leaf=5,
                            min_samples_split=7,
                            n_estimators=150,
                            n_jobs=4)
```

```python
rf.fit(X_train, Y_train)
```

```python
prediction = rf.predict(X_train)
cm = confusion_matrix(Y_train, prediction)
displ = ConfusionMatrixDisplay(cm)
displ.plot()
plt.show()
print(classification_report(Y_train, prediction))
```

```python
prediction = rf.predict(X_test)
cm = confusion_matrix(Y_test, prediction)
displ = ConfusionMatrixDisplay(cm)
displ.plot()
plt.show()
print(classification_report(Y_test, prediction))
```
